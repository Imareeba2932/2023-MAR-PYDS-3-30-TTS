Import library function

```python
from dputils.scrape import Tag, Scraper   #Tag and Scrapper are classes
```

- Tag class helps to decide what to collect from the page
- Scraper executes the scraping process 

```python
url = "https://timesofindia.indiatimes.com/news"
news_scraper = Scraper(url)
```

```python
Tag?
```

```python
Scraper?
```

```python
target = Tag(cls = 'listing4')
items = Tag('li')
title = Tag('span', cls = 'w_tle')
desc = Tag('span', cls = 'w_desc')
```

```python
news_scraper.get_all(target, items, heading=title, desc=desc)
```

```python
result = news_scraper.get_all(target, items, heading=title, desc=desc)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(result)
df
```

```python
url = ('https://www.amazon.in/s?k=laptop&adgrpid=1326012632290860&hvadid=82876047499036&hvbmt=be&hvdev=c&hvlocphy=163963&hvnetw=o&hvqmt=e&hvtargid=kwd-82876679612613%3Aloc-90&hydadcr=16601_1945050&tag=msndeskstdin-21&ref=pd_sl_4ghj4dv6zn_e')
```

```python
amazon_scraper = Scraper(url)
```

```python
target = Tag(cls = 's-main-slot s-result-list s-search-results sg-row')
items = Tag(cls = 's-result-item')
title = Tag('h2', cls='a-size-mini a-spacing-none a-color-base s-line-clamp-2')
price = Tag('span', cls='a-price')
```

```python
amazon_scraper.get_all(target, items, heading=title, price=price)
```

```python
result = amazon_scraper.get_all(target, items, heading=title, price=price)
```

```python
df=pd.DataFrame(result)
df
```
